```python
import numpy as np
import pandas as pd
import statistics as st
import random as rd
import matplotlib.pyplot as plt

# Setting seed for reproducibility
np.random.seed(25)

# Loading diabetes data
diabetes_data = pd.read_csv('/content/diabetes.csv')
df = pd.DataFrame(diabetes_data)

# Taking a sample of 25 data points
sample_data = diabetes_data.sample(n=25, random_state=35)

# Calculating mean glucose level of the sample
mean_of_glucose_sample = st.mean(sample_data.loc[:, "Glucose"])
print("Mean glucose level of the sample:", mean_of_glucose_sample)

# Finding maximum glucose level in the sample
max_of_glucose_sample = max(sample_data.loc[:, "Glucose"])
print("Maximum glucose level in the sample:", max_of_glucose_sample)

# Calculating mean glucose level of the entire population
mean_of_glucose_population = st.mean(df.loc[:, "Glucose"])
print("Mean glucose level of the population:", mean_of_glucose_population)

# Finding maximum glucose level in the entire population
max_of_glucose_population = max(df.loc[:, "Glucose"])
print("Maximum glucose level in the population:", max_of_glucose_population)

# Plotting a bar chart to compare mean glucose levels of sample and population
plt.bar(['Mean of Sample', 'Mean of Population'], [mean_of_glucose_sample, mean_of_glucose_population], color='blue')
plt.ylabel('Glucose Values')
plt.title('Comparison of Glucose Statistics: Sample vs. Population')
plt.show()
```

```python
# Creating a bar chart to compare the maximum glucose levels of the sample and population
plt.bar(['Max of Sample', 'Max of Population'], [max_of_glucose_sample, max_of_glucose_population], color='blue')
plt.ylabel('Glucose Values')
plt.title('Comparison of Glucose Statistics: Max of Sample vs. Population')
plt.show()
```

```python
# Setting seed for reproducibility
np.random.seed(25)

# Number of bootstrap samples and size of each sample
no_of_samples = 500
sample_size = 200

# Arrays to store bootstrap statistics
bootstrap_means = np.zeros(no_of_samples)
bootstrap_stds = np.zeros(no_of_samples)
bootstrap_percentiles = np.zeros(no_of_samples)

# Generating bootstrap samples and calculating statistics
for i in range(no_of_samples):
    bootstrap_sample = np.random.choice(df['BloodPressure'], size=sample_size, replace=True)
    bootstrap_means[i] = np.mean(bootstrap_sample)
    bootstrap_stds[i] = np.std(bootstrap_sample)
    bootstrap_percentiles[i] = np.percentile(bootstrap_sample, 63)

# Calculating population statistics
population_mean = np.mean(df['BloodPressure'])
population_std = np.std(df['BloodPressure'])
population_percentile = np.percentile(df['BloodPressure'], 63)

# Plotting histograms and population mean
plt.figure(figsize=(12, 6))
plt.hist(bootstrap_means, bins=25, color='red', alpha=0.5)
plt.axvline(population_mean, color='blue', linewidth=3)
plt.title('Comparison of Bootstrap Means and Population Mean')
```

```python
# Creating a figure for plotting
plt.figure(figsize=(12, 6))

# Plotting a histogram of bootstrap standard deviations
plt.hist(bootstrap_stds, bins=30, color='black', alpha=0.5)

# Adding a vertical line for the population standard deviation
plt.axvline(population_std, color='white', linewidth=3)

# Adding title to the plot
plt.title('Comparison of Bootstrap Standard Deviations and Population Standard Deviation')
```

```python
# Creating a figure for plotting
plt.figure(figsize=(12, 6))

# Plotting a histogram of bootstrap percentiles
plt.hist(bootstrap_percentiles, bins=30, color='red', alpha=0.5)

# Adding a vertical line for the population percentile
plt.axvline(population_percentile, color='blue', linewidth=3)

# Adding title to the plot
plt.title('Comparison of Bootstrap Percentiles and Population Percentile')
```
